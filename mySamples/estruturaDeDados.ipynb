##### Listas: Estrutura

```python
myList = [] #colchetes representam uma lista

print(type(myList))

myList = [
    'paulo'
    ,41
    ,['banana','maça','pera','uva']
    ,('tv',1000,'rosa') #tupla
    ,{11,13,15,17,19} # conjunto
    ,{1:'amanda',2:'antonio',3:'bianca',4:'bruno'}#dicionario
    ,'pietro'
    ,10.5
] #suporta vários tipos (é heterogênea)

print("\n# Listas são heterogêneas suportam vários tipos MISTURADOS na mesma lista")
for item in myList:
    print(f'tipo: {type(item)} \t| item: {item}')
```

##### Listas: Iterando

```python

print("\n# Interando a lista e os iteravies que ela possuir\nMÉTODO: 1(string são incluidos)")
for item in myList:
    try:
        iter(item)
        for i in item:
            print(i)
    except:
        print(item)
```

```python

print("\n# Interando a lista e os iteravies que ela possuir\nMÉTODO: 2(string NÃO são incluidos)")
for item in myList:
    try:
        iter(item)
        if not isinstance(item,str):
            for i in item:
                print(i)
        else:
            print(item)
    except:
        print(item)
```

##### Listas: Inserindo itens

```python
# Método APPEND - Insere no final da lista

print(f'A lista tem {len(myList)} itens')
print(myList)

print('\n INCLUINDO UM ITEM.....\n')

myList.append('Patrizia')

print(f'A lista tem {len(myList)} itens')
print(myList)
```

```python
# Método EXTEND - Incorpora uma outra lista na lista origem

mySecondList = ['pernalonga','patolino','gaguinho','piu-piu','frajola']

print(f'A lista tem {len(myList)} itens')
print(myList)

print(f'\nA lista SECUNDARIO tem {len(mySecondList)} itens')
print(mySecondList)

print('\n EXTENDENDO A LISTA.....\n')

myList.extend(mySecondList)

print(f'A lista tem {len(myList)} itens')
print(myList)
```

```python
# Método INSERT - Insere um item em uma posição especifica da lista

print(f'A lista tem {len(myList)} itens')
print(myList)

print('\n INSERINDO UM ITEM EM UM INDICE ESPECIFICO (neste caso no indice [0]).....\n')

myList.insert(0,'PIETRO')

print(f'A lista tem {len(myList)} itens')
print(myList)
```

##### Listas: Removendo itens

```python
# Método REMOVE - Remove itens baseado no VALOR do item

print(f'A lista tem {len(myList)} itens')
print(myList)

print('\n REMOVENDO O ITEM [PAULO]\n')

myList.remove('paulo')

print(f'A lista tem {len(myList)} itens')
print(myList)
```

```python
# Método POP - Remove itens baseado no INDICE

print(f'A lista tem {len(myList)} itens')
print(myList)

print('\n REMOVENDO O ITEM INDICE (2)\n')

myList.pop(2)

print(f'A lista tem {len(myList)} itens')
print(myList)

myList.pop() # Remove o ultimo item

print(f'\nA lista tem {len(myList)} itens')
print(myList)
```

##### Listas: Cópiando uma Lista

###### As listas como tudo são OBJETOS, se simplesmente atribuir de uma variavel para outra, não será uma cópia, é o acesso ao mesmo OBJETO por duas variaveis

```python
# Método COPY() - Cria uma cópia

# exemplo sem copy

listCopy = myList

print(f'A lista myList tem {len(myList)} itens')
print(myList)

print(f'\nA lista listCopy tem {len(listCopy)} itens')
print(listCopy)

print('\n REMOVENDO O ITEM INDICE (0) DA LISTA listCopy\n')

listCopy.pop(0)

print(f'\nA lista listCopy tem {len(listCopy)} itens')
print(listCopy)

print(f'\nA lista myList tem {len(myList)} itens, também foi afetada pela remoção feita na lista listCopy')
print(myList)

print('\n COPIANDO A LISTA myList para listCopy\n')
listCopy = myList.copy()

print('\n REMOVENDO O ITEM INDICE (0) DA LISTA listCopy\n')
listCopy.pop(0)

print(f'\nA lista listCopy tem {len(listCopy)} itens')
print(listCopy)

print(f'\nA lista myList tem {len(myList)} itens, NÃO foi afetada pela remoção feita na lista listCopy')
print(myList)
```

#### Listas: Convertendo em listas ( list() e join() )

```python
# Método list() convert um outro tipo em lista

#Dicionario
dicionario = {1:'a',2:'b',3:'c'}

print(f'\nConversão de dicionário')
print(f'\n{type(dicionario)}')
print(f'{dicionario}')

listDic = list(dicionario.values())

print(f'\n{type(listDic)}')
print(f'{listDic}')

#Conjunto
conjunto = {'a','b','c'}

print(f'\nConversão de conjunto')
print(f'\n{type(conjunto)}')
print(f'{conjunto}')

listConjunto = list(conjunto)

print(f'\n{type(listConjunto)}')
print(f'{listConjunto}')

#Tupla
tupla = ('a','b','c')

print(f'\nConversão de tupla')
print(f'\n{type(tupla)}')
print(f'{tupla}')

listTupla = list(tupla)

print(f'\n{type(listTupla)}')
print(f'{listTupla}')

#String
nome = 'Pietro'

print(f'\nConversão de string')
print(f'\n{type(nome)}')
print(f'{nome}')

listString = list(nome)

print(f'\n{type(listString)}')
print(f'{listString}')

# Criando uma String a partir de uma 
# Método join() une qualquer iterable onde o conteudo seja uma string

print(f'\nCriando STRING a partir de um iterable STR')
stringResult = ''.join(tupla) #iterable: Iterable[str]
print(stringResult)
```

### Dicionarios

```python
# INIT

def carregarDicionario()->dict:
    """Inicializa um dicionario para os estudos"""
    myDic = {
    "nome":"paulo",
    "idade":41,
    "altura":1.78,
    "olhos": "castanhos",
    "cabelo": "preto"
    }
    return myDic
```

##### Dicionarios: Estrutura

```python
myDic = carregarDicionario()

# Dicionários no python assim como as listas são Heterôgeneos
print(type(myDic))
F.imprimirLista(myDic)
```

##### Dicionarios: Iterando

```python
myDic = carregarDicionario()

print(f"Imprimindo o dicionário: {myDic}")

#CHAVE
print("\nDicionário iterado por chave (Padrão - não é necessário função)")
for k in myDic:
    print(f"chave: {k}")

#CHAVE/VALOR
print("\nDicionário iterado por chave (Necessário função item())")
for k,v in myDic.items():
    print(f"chave: {k}\t| valor: {v}")

#VALOR
print("\nDicionário iterado por chave (Necessário função values())")
for v in myDic.values():
    print(f"valor: {v}")
```

##### Dicionarios: Inserindo itens

```python
# Para inserir itens basta acessar o dicionário e atribuir um valor a uma chave

myDic = carregarDicionario()

print(myDic)

#Atribuindo
myDic["peso"] = 76

print("Adicionado a CHAVE peso")

print(myDic)
```

##### Dicionarios: Removendo itens

```python
#Removendo por chave
myDic = carregarDicionario()
print(myDic)

print("Removando a CHAVE cabelo")

result = myDic.pop("cabelo")

print(myDic)
print(f'valor removido: {result}')
```

```python
#Removendo o ultimo item

print(myDic)

print("Removando a CHAVE cabelo")

result = myDic.popitem()

print(myDic)
print(f'valor removido: {result}')
```

##### Dicionarios: Atribuindo um valr PADRÃO

```python
# Atribuindo um valor padrão para chaves sem valor

myDic = carregarDicionario()

print(myDic)

result = myDic.setdefault("calçado",41)

print(myDic)

print(f'retorno do setdefault(): {result}')
```
